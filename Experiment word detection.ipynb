```python
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import cv2
import os
import math


from htr import page_detection as pd

%matplotlib inline
plt.rcParams['figure.figsize'] = (13.0, 13.0)


def implt(img, cmp=None, t=''):
    plt.imshow(img, cmap=cmp)
    plt.title(t)
    plt.show()
```

```python
path = "./data/slova/s4.jpg"
image = pd.detect(path)
img = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
implt(img, cmp='gray')
```

```python
def sobel(channel):
    """ The Sobel Operator"""
    sobelX = cv2.Sobel(channel, cv2.CV_16S, 1, 0)
    sobelY = cv2.Sobel(channel, cv2.CV_16S, 0, 1)
    # Combine x, y gradient magnitudes sqrt(x^2 + y^2)
    sobel = np.hypot(sobelX, sobelY)
    sobel[sobel > 255] = 255
    return np.uint8(sobel)


def edge_detect(im):
    """ 
    Edge detection 
    The Sobel operator is applied for each image layer (RGB)
    """
    return np.max(np.array([sobel(im[:,:, 0]), sobel(im[:,:, 1]), sobel(im[:,:, 2]) ]), axis=0)

# Image pre-processing - blur, edges, threshold, closing
blurred = cv2.GaussianBlur(image, (5, 5), 18)
edges = edge_detect(blurred)
ret, edges = cv2.threshold(edges, 80, 255, cv2.THRESH_BINARY)
before_dilate = cv2.morphologyEx(edges, cv2.MORPH_CLOSE, np.ones((20,20), np.uint8))

kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (10, 10))
bw_image = cv2.dilate(before_dilate, kernel, iterations = 4)

implt(edges, 'gray', 'Sobel operator')
implt(bw_image, 'gray', 'Final closing')
```

```python
def rectangle_area(c):
    return c[2]*c[3]
```

```python
def reject_outliers(data, m=3):
    
    # Make histograms
    
    data[:,4] = (data[:,4] - np.amin(data[:,4]))/(np.amax(data[:,4]) - np.amin(data[:,4]))
    plt.plot(data[:,4], color='red')
    #plt.hist(data[:,4], bins=100)
    
    data = data[abs(data[:,4] - np.mean(data[:,4])) < m * np.std(data[:,4])]
    plt.plot(data[:,4], color='blue')
    
    red_patch = mpatches.Patch(color='red', label='Standardized data')
    blue_patch = mpatches.Patch(color='blue', label='Outliner cut data')
    plt.legend(handles=[red_patch, blue_patch])
    plt.show()
    return data
```

```python
def extract_and_save_chars(boxes, original_image):
    count = 0
    for (x, y, w, h) in boxes:
        roi = original_image[y:y+h, x:x+w]
        cv2.imwrite("./training_data/velika1/" + str(count) + ".jpg", roi)
        count += 1
```

```python
def text_detect(img, before_dilate):
    """ Text detection using contours """
    
    # Finding contours
    mask = np.zeros(img.shape, np.uint8)
    im2, cnt, hierarchy = cv2.findContours(np.copy(img), cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
    
    # Variables for contour index and words' bounding boxes
    index = 0    
    boxes = []
    
    boxes_with_area = np.empty((0,5), float)
    
    box_area_sum = 0
    
    # CCOMP hierarchy: [Next, Previous, First Child, Parent]
    # cv2.RETR_CCOMP - contours into 2 levels
    # Go through all contours in first level
    while (index >= 0):
        x,y,w,h = cv2.boundingRect(cnt[index])
        # Get only the contour
        cv2.drawContours(mask, cnt, index, (255, 255, 255), cv2.FILLED)
        maskROI = before_dilate[y:y+h, x:x+w]
        
        #implt(maskROI, t='Bounding rectangles')
        
        # Ratio of white pixels to area of bounding rectangle
        r = cv2.countNonZero(maskROI) / (w * h)
        
        # Limits for text (white pixel ratio, width, height)
        # TODO Test h/w and w/h ratios
        if r > 0.1 and 2000 > w > 10 and 1600 > h > 10 and h/w < 3 and w/h < 10:
            a = [x, y, w, h]
            area = rectangle_area(a);
            box_with_area = np.array([x, y, w, h, area])
            boxes_with_area = np.vstack((boxes_with_area, box_with_area))
        
        # Index of next contour
        index = hierarchy[0][index][0]
    
    boxes = reject_outliers(boxes_with_area)
    
    image_copy = image.copy()
    bounding_boxes = np.empty((0,4), int)
    for (x, y, w, h, i) in boxes:
        x = int(x)
        y = int(y)
        w = int(w)
        h = int(h)
        cv2.rectangle(image_copy, (x, y),(x+w,y+h), (0, 255, 0), 8)
        bounding_boxes = np.vstack((bounding_boxes, np.array([x, y, w, h])))

    implt(image_copy, t='Bounding rectangles')

    # Sort 
    def keyy(val):
        return math.sqrt(val[0]**2 + val[1]**2) + val[1]
    
    bounding_boxes = bounding_boxes.tolist()
    bounding_boxes.sort(key = keyy) 
    
    return bounding_boxes  
```
